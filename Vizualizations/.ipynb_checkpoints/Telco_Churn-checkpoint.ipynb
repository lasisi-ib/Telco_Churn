```python
fig, ax = plt.subplots(nrows =2, ncols = 3, figsize = (15,10))
hist('Total Charges', ax[0][0])
hist('Tenure Months', ax[0][1])
hist('Monthly Charges', ax[0][2])
hist('Churn Value', ax[1][0])
hist('Churn Score', ax[1][1])
hist('CLTV', ax[1][2])
```

I believe Monthly Charges variable will shed more light on how charges influence churn.</p>
<h3> Tenure(Months)</h3>

<h3>Monthly Charges</h3>

```python
def barplot(var, ax = None):
    temp = pd.DataFrame(pd.crosstab(df[var], df['Churn Label']))
    for i in range(len(temp)):
        temp.iloc[i] = temp.iloc[i] / temp.iloc[i].sum() * 100

    return temp.plot(kind= 'bar', stacked = True, ax = ax)
    
```

# Churn VS Categorical Variables

```python
def  stackedbar(var,axs, data=df):
    tomp = pd.DataFrame(pd.crosstab(df[var], data['Churn Label']))
    tomp1 = pd.DataFrame(pd.crosstab(df[var], data['Churn Label']))
    for i in range(len(tomp)):
        tomp1.iloc[i]= round((tomp.iloc[i]/ tomp.iloc[i].sum()) * 100, 0)
    
    tt = tomp.plot(kind ='bar', ax=axs, stacked= True)
    for d in tt.patches:
        r,c = indexer(d.get_height(), tomp)
        axs.annotate(f'{tomp1[tomp.columns[c]][r]}%',
                     ((d.get_x() + d.get_width() / 2) * 0.96, (d.get_y() + d.get_height() / 2 )* 0.99),
                    color = 'black')
   
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15,12))
stackedbar('Gender', ax[0][0])
stackedbar('Senior Citizen', ax[0][1])
stackedbar('Partner', ax[1][0])
stackedbar('Dependents', ax[1][1])
plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15,12))
stackedbar('Phone Service', ax[0][0])
stackedbar('Multiple Lines', ax[0][1])
stackedbar('Internet Service', ax[1][0])
stackedbar('Online Security', ax[1][1])
plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15,12))
stackedbar('Online Backup', ax[0][0])
stackedbar('Device Protection', ax[0][1])
stackedbar('Tech Support', ax[1][0])
stackedbar('Streaming TV', ax[1][1])
plt.tight_layout()
```
